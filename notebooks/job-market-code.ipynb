```python
def count_skills(dataframe):
    
    skill_count = {}

    for skills in dataframe["extracted_skills"]:
        for skill in skills:
            if skill in skill_count:
                skill_count[skill] += 1
            else:
                skill_count[skill] = 1

    return skill_count
```

```python
def create_skill_dataframe(skill_count):

    sorted_skills = sorted(
        skill_count.items(),
        key=lambda x:x[1],
        reverse=True
    )

    return pd.DataFrame(
        sorted_skills,
        columns=["skill","count"]
    )
```

```python
def filter_role(dataframe, role):

    return dataframe[
        dataframe["title"].str.contains(
            role,
            case=False,
            na=False
        )
    ]
```

```python
data_analyst_df = filter_role(df,"Data Analyst")

data_scientist_df = filter_role(df,"Data Scientist")

data_engineer_df = filter_role(df,"Data Engineer")
```

```python
analyst_skill_count=count_skills(data_analyst_df)
scientist_skill_count=count_skills(data_scientist_df)
engineer_skill_count=count_skills(data_engineer_df)
```

```python
analyst_df = create_skill_dataframe(
    analyst_skill_count
)

scientist_df = create_skill_dataframe(
    scientist_skill_count
)

engineer_df = create_skill_dataframe(
    engineer_skill_count
)
```

```python
def plot_bar_chart(dataframe, x_column, y_column, title,top_n=None):
    if top_n is not None:
      dataframe = dataframe.nlargest(top_n, x_column)
    plt.figure(figsize=(10,6))

    ax = sns.barplot(
        data=dataframe,
        x=x_column,
        y=y_column,
        color="steelblue"
    )

    plt.title(title, fontsize=16, fontweight="bold")

    plt.xlabel(x_column.capitalize())
    plt.ylabel(y_column.capitalize())

    for container in ax.containers:
        ax.bar_label(container, padding=5)

    plt.grid(axis="x", linestyle="--", alpha=0.5)

    plt.xlim(0, dataframe[x_column].max() * 1.1)

    plt.tight_layout()

    plt.show()
```

```python
plot_bar_chart(
    analyst_df,
    "count",
    "skill",
    "Top 10 Skills for Data Analyst",
    top_n=10
)
plot_bar_chart(
    scientist_df.nlargest(10, "count"),
    "count",
    "skill",
    "Top 10 Skills for Data Scientist",
    top_n=10
)
plot_bar_chart(
    engineer_df.nlargest(10, "count"),
    "count",
    "skill",
    "Top 10 Skills for Data Engineer",
    top_n=10
)
```

```python
skill_categories = {
    "Programming": [
        "python", "r"
    ],

    "Database": [
        "sql"
    ],

    "Cloud": [
        "aws", "azure", "google cloud"
    ],

    "Business Intelligence & Visualization": [
        "excel", "tableau", "power bi"
    ],

    "Machine Learning": [
        "machine learning",
        "deep learning",
        "scikit-learn",
        "tensorflow",
        "pytorch"
    ],

    "Data Analysis": [
        "pandas",
        "numpy",
        "statistics"
    ],

    "Big Data": [
        "spark",
        "hadoop"
    ],

    "DevOps": [
        "docker",
        "kubernetes",
        "linux"
    ],

    "Version Control": [
        "git"
    ]
}
```

```python
category_count = {}

for category, skills in skill_categories.items():
    category_count[category] = 0

    for skill in skills:
        if skill in skill_count:
            category_count[category] += skill_count[skill]
```

```python
skill_count
```

```python
category_count
```

```python
category_df = pd.DataFrame(
    list(category_count.items()),
    columns=["category", "count"]
)
category_df = category_df.sort_values(
    by="count",
    ascending=False
)
plot_bar_chart(
    category_df,
    "count",
    "category",
    "Most In-Demand Skill Categories"
)
```
