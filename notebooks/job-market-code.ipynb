```python
skills = [
    "python",
    "sql",
    "excel",
    "power bi",
    "tableau",
    "aws",
    "azure",
    "google cloud",
    "pandas",
    "numpy",
    "scikit-learn",
    "tensorflow",
    "pytorch",
    "machine learning",
    "deep learning",
    "statistics",
    "spark",
    "hadoop",
    "docker",
    "kubernetes",
    "git",
    "linux"
]
all_skills = []
df["description_clean"] = df["description_clean"].fillna("")
for i in range(len(df)):
    current_job_skills = []
    description=df["description_clean"].iloc[i]
    for skill in skills:
        if skill in description:
            current_job_skills.append(skill)
    all_skills.append(current_job_skills)
```

```python
skill_count = {}

for skills in df["extracted_skills"]:
    for skill in skills:
        if skill in skill_count:
            skill_count[skill] += 1
        else:
            skill_count[skill] = 1
```

```python
list(skill_count.items())[:10]
```

```python
sorted_skills = sorted(
    skill_count.items(),
    key=lambda x: x[1],
    reverse=True
)
```

```python
print(sorted_skills)
```

```python
skill_df = pd.DataFrame(
    sorted_skills,
    columns=["skill", "count"]
)
```

```python
skill_df
```

```python
top10 = skill_df.head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select Top 10 skills
top10 = skill_df.head(10)

# Create figure
plt.figure(figsize=(10,6))

# Create horizontal bar chart
ax = sns.barplot(
    data=top10,
    x="count",
    y="skill",
    color="steelblue"
)

# Add title and labels
plt.title("Top 10 Most Demanded Skills", fontsize=16, fontweight="bold")
plt.xlabel("Number of Job Postings", fontsize=12)
plt.ylabel("Skill", fontsize=12)

# Add value labels at the end of each bar
for container in ax.containers:
    ax.bar_label(container, fmt="%d", padding=5)

# Add grid on x-axis
plt.grid(axis="x", linestyle="--", alpha=0.5)

# Save the figure
plt.savefig("top10_skills.png", dpi=300, bbox_inches="tight")

# Adjust layout
plt.tight_layout()

# Display plot
plt.show()
```
